```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("GoDash - percentileMOSandResolutionsHost1.csv")
```

```python
df.head()
```

```python
data_5MB = df[(df['Bandwidth'] == 5)]
data_10MB = df[(df['Bandwidth'] == 10)]
data_15MB = df[(df['Bandwidth'] == 15)]
```

```python
bestCaseArbiter = df[(df['Delay'] == 10) & (df['Algorithms'] == 'arbiter')]
```

```python
worstCaseArbiter = df[(df['Delay'] == 100)& (df['Algorithms'] == 'arbiter')]
```

```python
bestCaseConv = df[(df['Delay'] == 10) & (df['Algorithms'] == 'conv')]
```

```python
worstCaseConv = df[(df['Delay'] == 100) & (df['Algorithms'] == 'conv')]
```

```python
g = sns.FacetGrid(bestCaseArbiter, col="Bandwidth", hue="Competition",hue_kws={"marker": ["*", "o","s"]})
g.map(plt.scatter, "MOS", "Percentile", alpha=.7)
g.add_legend(title= 'Nodes');
g.set_axis_labels("Mean Opinion Score", "Percentile")
axes = g.axes.flatten()
axes[0].set_title("5 mbit/s")
axes[1].set_title("10 mbit/s")
axes[2].set_title("15 mbit/s")
plt.subplots_adjust(top=0.80)
g.fig.suptitle('Best Case: Delay 10ms & Node = 1') # can also get the figure from plt.gcf()
```

```python
g = sns.FacetGrid(worstCaseArbiter, col="Bandwidth", hue="Competition",hue_kws={"marker": ["*", "o","s"]})
g.map(plt.scatter, "MOS", "Percentile", alpha=.7)
g.add_legend(title= 'Nodes');
g.set_axis_labels("Mean Opinion Score", "Percentile")
axes = g.axes.flatten()
axes[0].set_title("5 mbit/s")
axes[1].set_title("10 mbit/s")
axes[2].set_title("15 mbit/s")
plt.subplots_adjust(top=0.80)
g.fig.suptitle('Worst Case: Delay 100ms & Node = 1') # can also get the figure from plt.gcf()
```
