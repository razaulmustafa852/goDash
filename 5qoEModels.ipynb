```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("datasOld/GoDash - allQoEExp28.csv")
```

```python
df.head()
```

```python
fig, axs = plt.subplots(5,figsize=(7,5.5))
fig.suptitle('Arbiter: 15 mbit/s, delay 10ms and users 5, 1st user QoE')
axs[0].plot(df['intSeg'], df['P1203'],'tab:orange')
axs[0].set(ylabel='P1203')
axs[1].plot(df['intSeg'],df['Clae'],'tab:green')
axs[1].set(ylabel='Clae')
axs[2].plot(df['intSeg'],df['Duanmu'],'tab:red')
axs[2].set(ylabel='Duanmu')
axs[3].plot(df['intSeg'],df['Yin'],'tab:blue')
axs[3].set(ylabel='Yin')
axs[4].plot(df['intSeg'],df['Yu'],'tab:brown')
axs[4].set(ylabel='Yu',xlabel='Segments')
#for ax in axs.flat:
       #ax.set(xlabel='Segments', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
       ax.label_outer()
plt.savefig('qoe5Models.eps', format='eps')
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.scatter(df['intSeg'],df['Resolutions'])

ax.set(xlabel='Segments', ylabel='Resolutions',
       title='Arbiter: Segments in each resolutions')
ax.grid()


plt.savefig('Resolutions5Models.eps', format='eps')
```

```python


fig, axs = plt.subplots(5,figsize=(6,8))
fig.suptitle('Arbiter: 15 mbit/s, delay 10ms and users 5, 1st user QoE')
axs[0].plot(df['intSeg'], df['P1203'],'tab:orange')
axs[0].set(ylabel='P1203')
axs[1].plot(df['intSeg'],df['Clae'],'tab:green')
axs[1].set(ylabel='Clae')
axs[2].plot(df['intSeg'],df['Duanmu'],'tab:red')
axs[2].set(ylabel='Duanmu')
axs[3].plot(df['intSeg'],df['Yin'],'tab:blue')
axs[3].set(ylabel='Yin')
axs[4].plot(df['intSeg'],df['Yu'],'tab:brown')
axs[4].set(ylabel='Yu',xlabel='Segments')
#for ax in axs.flat:
       #ax.set(xlabel='Segments', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
       ax.label_outer()
ax0 = axs[0].twinx() 
ax1 =axs[1].twinx()  
ax2 =axs[2].twinx()  
ax3 =axs[3].twinx()  
ax4 =axs[4].twinx() 


ax0.plot(df['intSeg'], df['Resolutions'],'tab:brown')
ax1.plot(df['intSeg'], df['Resolutions'],'tab:blue')
ax2.plot(df['intSeg'], df['Resolutions'],'tab:pink')
ax3.plot(df['intSeg'], df['Resolutions'],'tab:green')
ax4.plot(df['intSeg'], df['Resolutions'],'tab:orange')
#plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0.1)
plt.savefig('bothYModelsandResolutionsQoE.eps', format='eps')
```
