```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Percentile109050 - Sheet2.csv")
```

```python
df.head()
```

```python
#mean_10 = df[(df['Delay'] == 10)]
```

```python
arbiter_5 = df[(df['Algorithms'] == 'arbiter') & (df['Competition'] == 5)]
arbiter_10 = df[(df['Algorithms'] == 'arbiter') & (df['Competition'] == 10)]
arbiter_15 = df[(df['Algorithms'] == 'arbiter') & (df['Competition'] == 15)]
```

```python
conv_5 = df[(df['Algorithms'] == 'conv') & (df['Competition'] == 5)]
conv_10 = df[(df['Algorithms'] == 'conv') & (df['Competition'] == 10)]
conv_15 = df[(df['Algorithms'] == 'conv') & (df['Competition'] == 15)]
```

```python
elastic_5 = df[(df['Algorithms'] == 'elastic') & (df['Competition'] == 5)]
elastic_10 = df[(df['Algorithms'] == 'elastic') & (df['Competition'] == 10)]
elastic_15 = df[(df['Algorithms'] == 'elastic') & (df['Competition'] == 15)]
```

```python
g = sns.FacetGrid(elastic_5, col="Delay", hue="Percentile",hue_kws={"marker": ["v", "o","^"]})
g.map(plt.scatter, "Bandwidth", "MOS")
#g.add_legend(title= 'Nodes');
g.set_axis_labels("Bandwidth", "Mean Opinion Score")
g.add_legend(title= 'Percentile')
plt.xticks([5, 10, 15])


axes = g.axes.flatten()
axes[0].set_title("Delay 10 ms")
axes[1].set_title("Delay 50 ms")
axes[2].set_title("Delay 100 ms")
plt.subplots_adjust(top=0.80)
g.fig.suptitle('Elastic: 5 Nodes competing for video') # can also get the figure from plt.gcf()
#plt.savefig('elastic.eps', format='eps')
```
